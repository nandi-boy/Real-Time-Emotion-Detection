```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import os
import cv2
import pickle # help to zip/compress the image file in byte(binary)
```

```python
# accessing the images present inside 'Emotions'
# getcwd :- fetching the current working directory 
img_dir = os.path.join(os.getcwd(),'Emotion')
img_dir = r'C:\Users\AYAN NANDI\OneDrive\Desktop\AIML-Training\Emotion\Emotion'
```

```python
print(os.getcwd())
```

```python
# creating a path to store the pickle files inside the data folder
img_final = os.path.join(os.getcwd(),'data')
img_final = r"C:\Users\AYAN NANDI\OneDrive\Desktop\AIML-Training\Machine Learning\Real Time Emotion detection\data"
```

```python
# to store all the inputs(images) and outputs(labels) in different []
images = []
labels = []
```

```python
# performing image preprocessing and saving the files 
# listdir: to list all folders/files present in any directory 
for i in os.listdir(img_dir):
    for j in os.listdir(os.path.join(img_dir,i)):  # to iterate over 'Emotions' folder
        img_path = os.path.join(os.path.join(img_dir,i),j) # to iterate over sub folders ['angry','sad'] etc
        img = cv2.imread(img_path) # to read the image
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # to convert 3 channels(BGR) to 1 channel (Grayscale)
        img = cv2.resize(img,(48,48)) # as well images have the shape of 48x48
        images.append(img)  # to store processed images in the 'images' list
        labels.append(i.split(' ')[0]) # to store the file name (i.e. sad,angry etc)
    print(i,'---> Completed')
```

```python
set(labels)
```

```python
images = np.array(images) 
labels = np.array(labels) 
```

```python
# Pickling the images and labels files 
# 'wb' - write binary as pickle file store only byte data 
# 'images.p' -> 'images' -> files name, '.p' -> pickle file's extension 
with open(os.path.join (img_final,'images.p'),'wb') as f:
    pickle.dump(images,f)

with open(os.path.join (img_final,'labels.p'),'wb') as f:
    pickle.dump(labels,f)
```
