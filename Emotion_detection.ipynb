```python
import pickle
import numpy as np
import tensorflow
import keras
```

```python
with open("/content/images.p",'rb') as f:
    x = pickle.load(f)

with open("/content/labels.p",'rb') as f:
    y = pickle.load(f)
```

```python
print ("Images : ",x.shape)
print ("Labels : ",y.shape)
```

```python
# Label Encoding Output
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
# Standardizing the images for CNN model :
x = x/255
x = x.reshape(-1,48,48,1)
```

```python
# Using one-hot encoding to output col
from keras.utils import to_categorical
y = to_categorical(y)
```

```python
# Splitting trainning and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
print ("Xtrain :",x_train.shape)
print ("Ytrain :",y_train.shape)
print ("Xtest :",x_test.shape)
print ("Ytest :",y_test.shape)
```

```python
# CNN Model training:
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten
```

```python
model = Sequential()
# 1st Convolutional Layer:
model.add(Conv2D(49, (3,3), activation='relu', input_shape=(48,48,1)))
model.add(MaxPooling2D(pool_size=(2,2)))

# 2nd Convolutional Layer:
model.add(Conv2D(69, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# 3rd Convolutional Layer:
model.add(Conv2D(123, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# Adding flatten layer: converts 2D to 1D
model.add(Flatten())

# Adding hidden layer:
model.add(Dense(240, activation='relu'))

#output layer:
model.add(Dense(5, activation='softmax'))
model.compile(Adam(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
# To save the model for deployment :
# .h5 -> HDF5 File format
# It stores models optimizer,s state, model architecture, training configuration,weights
model.save('Emotion_detection_model.h5')
print("Model Saved")
```
